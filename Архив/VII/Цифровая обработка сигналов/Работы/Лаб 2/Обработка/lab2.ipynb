### Пункт 1 (5)

```python
H_3 = lambda f: (np.sin(3*math.pi*f))/(3*np.sin(math.pi*f))
H_5 = lambda f: (np.sin(5*math.pi*f))/(5*np.sin(math.pi*f))
H_7 = lambda f: (np.sin(7*math.pi*f))/(7*np.sin(math.pi*f))
H_9 = lambda f: (np.sin(9*math.pi*f))/(9*np.sin(math.pi*f))
H_diff = lambda f: (1j*np.sin(2*math.pi*f)).imag
H_rect = lambda f: 1/(2j*np.sin(math.pi*f)).imag
H_trap = lambda f: (np.cos(math.pi*f))/(2j*np.sin(math.pi*f)).imag
H_simp = lambda f: (np.cos(2*math.pi*f)+2)/(3j*np.sin(math.pi*f)).imag
t = np.linspace(0.001, 0.5, 100)
```

```python
df = pd.DataFrame({'t': t, 'H_3': H_3(t), 'H_5': H_5(t), 'H_7': H_7(t), 'H_9': H_9(t)})
df_log = pd.DataFrame({'t': t, 'H_3': 20*np.log10(np.abs(H_3(t))), 'H_5': 20*np.log10(np.abs(H_5(t))),
                       'H_7': 20*np.log10(np.abs(H_7(t))), 'H_9': 20*np.log10(np.abs(H_9(t)))})
```

```python
df2 = df.melt(id_vars='t', var_name='cols', value_name='vals')
df2_log = df_log.melt(id_vars='t', var_name='cols', value_name='vals')
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df2, x='t', y='vals', hue='cols', linewidth=1.7,
                 kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels(r'$f$', r'$\tildeH(f)$')
plt.savefig('pics/1.png')
plt.show()

ax = sns.relplot(data=df2_log, x='t', y='vals', hue='cols', linewidth=1.7,
                 kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels(r'$f$', r'$\tildeH(f), Дб$')
plt.savefig('pics/5_1.png')
plt.show()
```

### Пункт 7

```python
t = np.linspace(0.001, 0.5, 100)
```

```python
H_r = lambda f: (1 / (2j*np.sin(math.pi*f))).imag
H_t = lambda f: (np.cos(math.pi*f) / (2j*np.sin(math.pi*f))).imag
H_s = lambda f: ((np.cos(2*math.pi*f)+2)/ (3j*np.sin(2*math.pi*f))).imag
```

```python
K_r = lambda f: math.pi * f / (np.sin(math.pi * f))
K_t = lambda f: np.cos(math.pi * f)*(math.pi*f/np.sin(f*math.pi))
K_s = lambda f: ((np.cos(2*math.pi*f) + 2)/3) * ((2*math.pi*f)/(np.sin(2*math.pi*f)))
```

```python
df_H = pd.DataFrame({'t': t, 'H_трапеция': H_t(t), 'H_Симпсон': H_s(t), 'H_прямоугольник': H_r(t)})
df_K = pd.DataFrame({'t': t, 'K_трапеция': K_t(t), 'K_Симпсон': K_s(t), 'K_прямоугольник': K_r(t)})
```

```python
df2_H = df_H.melt(id_vars='t', var_name='cols', value_name='vals')
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df2_H, x='t', y='vals', hue='cols', linewidth=1.7,
                 kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels(r'$f$', r'$\tildeH(f)$')
ax.set(ylim=(-4, 1))
plt.savefig('pics/6.png')
plt.show()
```

```python
df2_K = df_K.melt(id_vars='t', var_name='cols', value_name='vals')
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df2_K, x='t', y='vals', hue='cols', linewidth=1.7, style='cols',
                 kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels(r'$f$', r'$K(f)$')
ax.set(ylim=(0, 2))
plt.savefig('pics/7.png')
plt.show()
```
