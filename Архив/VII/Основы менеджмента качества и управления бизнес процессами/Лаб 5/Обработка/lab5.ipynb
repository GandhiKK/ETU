```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({'Тип дефектов': ['А', 'Б', 'В', 'Г', 'Д', 'Прочие'],
                  'Число дефектов': [54, 33, 22, 19, 13, 7],
                  'Накопленный процент от общего числа дефектов': [37, 60, 75, 88, 97, 100]})
```

```python
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df, x='Тип дефектов', y='Накопленный процент от общего числа дефектов',
                 kind='line', height=8.27, aspect=11.7/8.27, linewidth=2.5, color='r')
plt.bar(df['Тип дефектов'], df['Число дефектов'])
ax.set_axis_labels('Тип дефектов', 'Число дефектов')

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']-1.1, point['y']+1.8, str(point['val']))

label_point(pd.Series([1,2,3,4,5,6]), df['Накопленный процент от общего числа дефектов'],
            df['Накопленный процент от общего числа дефектов'], plt.gca())

plt.savefig('pics/1.png')
plt.show()
```

```python
df2 = pd.DataFrame({'Тип дефектов': ['Б', 'В', 'Г', 'Д', 'Прочие', 'А'],
                  'Число дефектов': [31, 24, 22, 14, 6, 5],
                  'Накопленный процент от общего числа дефектов': [31, 55, 76, 90, 96, 100]})
```

```python
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df2, x='Тип дефектов', y='Накопленный процент от общего числа дефектов', 
                 kind='line', height=8.27, aspect=11.7/8.27, linewidth=2.5, color='r')
plt.bar(df2['Тип дефектов'], df2['Число дефектов'])
ax.set_axis_labels('Тип дефектов', 'Число дефектов')
label_point(pd.Series([1,2,3,4,5,6]), df2['Накопленный процент от общего числа дефектов'], 
            df2['Накопленный процент от общего числа дефектов'], plt.gca())
plt.savefig('pics/2.png')
plt.show()
```
