```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

figsize = (12, 6)
dpi = 600
plt.style.use('bmh')
```

```python
data = pd.DataFrame({'num': [1, 2, 3, 4, 5, 6, 7, 8, 9],
                     'score': [64, 75, 79, 60, 81, 97, 113, 96, 108],
                     'title': [
                     'Обслуживание', 'Доп. услуги', 'Покупка фильмов',
                     'Оборудование', 'Афиша', 'Формирование цен',
                     'Мониторинг конкурентов', 'Прокат (Продажи)',
                     'Обслуживание посетителей'
]})
data
```

```python
med = np.mean(data['score'])
med
```

```python
plt.figure(figsize=figsize, dpi=dpi)
sns.scatterplot(data=data, x='num', y='score')
# plt.text(x=data['num'], y=data['score'], s=data['title'], cex=0.7)
for i in range(data.shape[0]):
    plt.text(x=data['num'][i] - 0.7, y=data['score'][i] + 1.6, s=f"{data['title'][i]} ({data['score'][i]})")
plt.text(2.5, 105, 'Перебор', c='r')
plt.text(2.5, 70, 'Не важно', c='r')
plt.text(7.0, 105, 'Норма', c='r')
plt.text(6.5, 70, 'Необходимы улучшения', c='r')
plt.hlines(med, 1, 9, colors='b', linestyles='--')
plt.vlines(5, min(data['score']), max(data['score']), colors='b', linestyles='--')
plt.xlabel('Важность')
plt.ylabel('Текущий показатель')
plt.title('Матрица показателей')
plt.savefig('mat_pok.png')
plt.show()
```
