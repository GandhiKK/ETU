```python
data1 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0.2, 0.6, 0.4, 0.6, 1, 1],
                     'score 2': [0, 0.6, 0.6, 0.2, 0.8, 0.8],
                     'score 3': [0.6, 0.6, 0.8, 0.8, 0.8, 1]
                      })

data2 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0.2, 0.8, 0.8, 0.8, 0.8, 1],
                     'score 2': [0.4, 1, 0.6, 0.6, 1, 1],
                     'score 3': [0.2, 0.4, 0.4, 0.8, 1, 1]
                      })

data3 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0, 0.4, 0.2, 0.4, 1, 1],
                     'score 2': [0, 0.4, 0.2, 0.6, 0.6, 0.8],
                     'score 3': [0, 0, 0.6, 0.4, 0.8, 0.6]
                      })

data4 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0.4, 1, 1, 1, 1, 1],
                     'score 2': [0.2, 1, 1, 1, 1, 1],
                     'score 3': [0.2, 0.8, 0.8, 1, 1, 0.8]
                      })

data5 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0.2, 0.4, 0.2, 0.4, 0.6, 0.6],
                     'score 2': [0.2, 0.2, 0.4, 0.4, 0.4, 0.8],
                     'score 3': [0.2, 0.4, 0.6, 0.4, 0.6, 0.8]
                      })

data6 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0, 0.2, 0.6, 0, 0, 0.4],
                     'score 2': [0, 0, 0.2, 0.4, 0.6, 0.4],
                     'score 3': [0, 0.2, 0.2, 0, 0.4, 0.4]
                      })

data7 = pd.DataFrame({'P(t)': [55, 110, 165, 220, 275, 330],
                     'score 1': [0.4, 0.6, 0.6, 0.6, 1, 1],
                     'score 2': [0.4, 0.4, 1, 0.6, 0.8, 1],
                     'score 3': [0.2, 0.6, 1, 0.6, 0.8, 0.6]
                      })
```

```python
def graph(cur_data, cur_title, cur_pic_title):
    plt.figure(figsize=figsize, dpi=dpi)
    new_data = pd.melt(cur_data, ["P(t)"])
    sns.lineplot(data=new_data, x="P(t)", y="value", hue='variable',
                 style="variable", palette="rocket_r")
    plt.xlabel('Время экспозиции, с')
    plt.ylabel('Значение')
    plt.legend(['Испытуемый 1', 'Испытуемый 2', 'Испытуемый 3'])
    plt.title(cur_title)
    plt.savefig('pics/' + cur_pic_title)
    plt.show()
```

```python
all_data = np.array([[data1, data2, data3, data4, data5, data6, data7],
                     ['Первый опыт', 'Второй опыт', 'Третий опыт', 'Четвертый опыт',
                      'Пятый опыт', 'Шестой опыт', 'Седьмой опыт'],
                     ['data1.png', 'data2.png', 'data3.png', 'data4.png',
                      'data5.png', 'data6.png', 'data7.png']], dtype=object)

for i in range(7):
    graph(all_data[0][i], all_data[1][i], all_data[2][i])
```
