```python
import numpy as np
import cv2 as cv
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = "all"
import requests 
import shutil 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from sklearn.cluster import KMeans
```

## Функции

```python
def viewImage(image, name_of_window):
    cv.namedWindow(name_of_window, cv.WINDOW_NORMAL)
    cv.imshow(name_of_window, image)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def pic_from_host(cur_path):
    image_url = 'https://i.ibb.co/cTB1B6G/T-Hn-Olskfn-A.jpg'
    filename = cur_path
    r = requests.get(image_url, stream = True)
    if r.status_code == 200:
        r.raw.decode_content = True
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)        
        print('Успешно загружено: ',filename)
    else:
        print('Не найдено')
```

```python
def centroid_histogram(clt):
    numLabels = np.arange(0, len(np.unique(clt.labels_)) + 1)
    (hist, _) = np.histogram(clt.labels_, bins=numLabels)
    hist = hist.astype("float")
    hist /= hist.sum()
    return hist
```

```python
def plot_colors(hist, centroids):
    bar = np.zeros((50, 300, 3), dtype="uint8")
    startX = 0
    for (percent, color) in zip(hist, centroids):
        endX = startX + (percent * 300)
        cv.rectangle(bar, (int(startX), 0), (int(endX), 50),
                      color.astype("uint8").tolist(), -1)
        startX = endX
    return bar
```

### Распределение

```python
def segments(K, iterations, iteration, img):
    for k in K:
        print(f'Количество кластеров: {k}')
        image = img.reshape((img.shape[0] * img.shape[1], 3))
        clt = KMeans(n_clusters = k, n_init = iterations, max_iter = iteration)
        clt.fit(image)
        hist = centroid_histogram(clt)
        bar = plot_colors(hist, clt.cluster_centers_)

        fig = plt.figure()
        ax = fig.add_subplot(211)
        ax.imshow(img)
        ax = fig.add_subplot(212)
        ax.imshow(bar)
        plt.show()
```

### Сегментация

```python
def main_seg(img, K):
    criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    Z = np.float32(img.reshape((-1,3)))
    for k in K:
        print(f'Количество кластеров: {k}')
        ret, label, center = cv.kmeans(Z, k, None, criteria, 10, cv.KMEANS_RANDOM_CENTERS)
        center = np.uint8(center)
        clusters = center[label.flatten()]
        result = clusters.reshape((img.shape))
        cv.imwrite('pics/result.bmp', cv.cvtColor(result, cv.COLOR_BGR2RGB))
        plt.imshow(result)
        plt.show()
```
