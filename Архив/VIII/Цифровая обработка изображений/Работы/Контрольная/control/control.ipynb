```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
i = np.arange(0,256,1)
```

```python
df = pd.DataFrame({'i':i})
```

```python
sns.set_theme(style="whitegrid", palette='deep', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df, x='i', y='i', kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels('X', 'Y')
plt.tight_layout()
plt.savefig('pics/1.png')
```

```python
df['e'] = [255*((np.arange(0,df.iloc[i,0]+1,1).sum())/(np.arange(0,256,1).sum())) for i in range(256)]
```

```python
ax = sns.relplot(data=df, x='i', y='e', kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels('X', 'Y')
plt.tight_layout()
plt.savefig('pics/2.png')
```

```python
ax = sns.relplot(x=df.i, y=255-df.i, kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels('X', 'Y')
plt.tight_layout()
plt.savefig('pics/3.png')
```
