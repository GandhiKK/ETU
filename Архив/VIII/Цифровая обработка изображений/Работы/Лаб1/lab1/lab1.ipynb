## Функции

### Вывод изображения

```python
def viewImage(image, name_of_window):
    cv.namedWindow(name_of_window, cv.WINDOW_NORMAL)
    cv.imshow(name_of_window, image)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

### Поворот

```python
def rotate_image(image, angle):
    image_size = (image.shape[1], image.shape[0])
    image_center = tuple(np.array(image_size) / 2)

    # конверитурем матрицу поворота
    rotation_M = np.vstack([cv.getRotationMatrix2D(image_center, angle, 1.0), 
                        [0, 0, 1]])
    rotation_M_upd = np.matrix(rotation_M[:2,:2])
    image_w2, image_h2  = image_size[0]/2.0, image_size[1]/2.0

    # повернутые координаты углов
    rotated_coords = [(np.array([-image_w2,  image_h2])*rotation_M_upd).A[0],
                      (np.array([ image_w2,  image_h2])*rotation_M_upd).A[0],
                      (np.array([-image_w2, -image_h2])*rotation_M_upd).A[0],
                      (np.array([ image_w2, -image_h2])*rotation_M_upd).A[0]]

    # размер нового изображения
    x_coords = [pt[0] for pt in rotated_coords]
    x_pos = [x for x in x_coords if x > 0]
    x_neg = [x for x in x_coords if x < 0]
    y_coords = [pt[1] for pt in rotated_coords]
    y_pos = [y for y in y_coords if y > 0]
    y_neg = [y for y in y_coords if y < 0]

    right_bound, left_bound = max(x_pos), min(x_neg)
    top_bound, bot_bound = max(y_pos), min(y_neg)

    new_w = int(abs(right_bound-left_bound))
    new_h = int(abs(top_bound-bot_bound))    
    # матрица сдвига для центрирования изображения
    trns_M = np.matrix([[1, 0, int(new_w/2.0-image_w2)],
                           [0, 1, int(new_h/2.0-image_h2)],
                           [0, 0, 1]])
    # поворот + сдвиг
    af_M = (np.matrix(trns_M)*np.matrix(rotation_M))[0:2, :]
    res = cv.warpAffine(image, af_M, (new_w, new_h),
                           flags=cv.INTER_LINEAR)
    return res
```

### Прямоугольник максимальной площади

```python
def largest_rotated_rect(w, h, angle):
    length = h if w < h else w
    qua = int(math.floor(angle / (math.pi / 2))) & 3
    sign = angle if ((qua & 1) == 0) else math.pi - angle
    alpha = (sign % math.pi + math.pi) % math.pi
    bb_w = w*math.cos(alpha) + h*math.sin(alpha)
    bb_h = w*math.sin(alpha) + h*math.cos(alpha)
    gamma = math.atan2(bb_w, bb_h) if (w < h) else math.atan2(bb_h, bb_w) 
    a = length * math.cos(alpha) * math.sin(alpha) / math.sin(math.pi-alpha-gamma)
    y = a * math.cos(gamma)
    x = y * math.tan(gamma)    
    wr, hr = bb_w-2*x, bb_h-2*y
    return (wr,hr)
```

### Обрезание вокруг центра

```python
def crop_around_center(image, width, height):
    image_size = (image.shape[1], image.shape[0])
    image_center = (int(image_size[0]/2.0), int(image_size[1]/2.0))
    width = image_size[0] if width > image_size[0] else width
    height = image_size[1] if height > image_size[1] else height
    x1, x2 = int(image_center[0] - width/2.0), int(image_center[0] + width/2.0)
    y1, y2 = int(image_center[1] - height/2.0), int(image_center[1] + height/2.0)
    return image[y1:y2,x1:x2]
```

### Обрезание

```python
def crop(cur_img, interpol, k):
    inter = {0: cv.INTER_NEAREST, 1: cv.INTER_LINEAR, 2: cv.INTER_CUBIC}
    h, w = cur_img.shape[:2]
    res = cv.resize(cur_img, (int(w*k), int(h/k)), 
                    inter[interpol])
    cv.imwrite('pics/4_result.bmp', res)
    return res
```

## Основа

```python
def pic_from_host(cur_path):
    image_url = 'https://i.ibb.co/cTB1B6G/T-Hn-Olskfn-A.jpg'
    filename = cur_path
    r = requests.get(image_url, stream = True)
    if r.status_code == 200:
        r.raw.decode_content = True
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)        
        print('Успешно загружено: ',filename)
    else:
        print('Не найдено')
```

```python
def_alpha = random.randrange(0, 360)
eps = 1e-10
def_k = round(random.random()*(1-eps)+eps,3)
def_interpl = 1
```

```python
def processing(path='pics/original.jpg', alpha=def_alpha, k=def_k, intrpl=def_interpl):
    print(f'Угол поворота alpha равен {alpha}, k равен {k}')
    img = cv.imread('pics/original.jpg')
    h, w = img.shape[:2]

    image_orig = np.copy(img)
    image_rotated = rotate_image(img, alpha)
    image_rotated_cropped = crop_around_center(image_rotated,
                           *largest_rotated_rect(w, h, math.radians(alpha)))

    res = crop(image_rotated_cropped, intrpl, k)

    cv.imwrite('pics/1_orig.jpg', image_orig)
    cv.imwrite('pics/2_rotated.jpg', image_rotated)
    cv.imwrite('pics/3_cropped.jpg', image_rotated_cropped)
    cv.imwrite('pics/result.bmp', res)
    
    plt.imshow(cv.cvtColor(image_orig, cv.COLOR_BGR2RGB))
    plt.show()
    plt.imshow(cv.cvtColor(image_rotated, cv.COLOR_BGR2RGB)) 
    plt.show()
    plt.imshow(cv.cvtColor(image_rotated_cropped, cv.COLOR_BGR2RGB)) 
    plt.show()
    plt.imshow(cv.cvtColor(res, cv.COLOR_BGR2RGB)) 
    plt.show()
```
