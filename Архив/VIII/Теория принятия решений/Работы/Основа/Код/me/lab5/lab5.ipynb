```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = "all"
import random
```

```python
df_abs = pd.read_excel('data/dabs2.xlsx', header=None)
df_abs = np.abs(df_abs)
df_otn = pd.read_excel('data/dotn2.xlsx', header=None)
df_otn = np.abs(df_otn)
```

```python
df_abs.columns = ['ax', 'ay', 'az']
df_otn.columns = ['rx', 'ry', 'rz']
df_abs['case'] = np.arange(1,31,1)
df_otn['case'] = np.arange(1,31,1)
```

```python
t = df_abs['ax'].values
t2 = df_otn['rx'].values
random.shuffle(t)
random.shuffle(t2)
df_abs['ax'] = t
df_otn['rx'] = t2
```

```python
# df_abs.iloc[:,[3,0]]
# df_otn.iloc[:,[3,0]]
```

```python
sns.set_theme(style="darkgrid", palette='rocket', context='notebook', font_scale=1.3)
ax = sns.relplot(data=df_abs, x='case', y='ax', kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels('Испытания', 'Абсолютная погрешность')
ax.fig.suptitle('Зависимость абсолютной погрешности')
plt.tight_layout()
plt.savefig('pics/abs_error.png')
```

```python
ax = sns.relplot(data=df_otn, x='case', y='rx', kind='line', height=8.27, aspect=11.7/8.27)
ax.set_axis_labels('Испытания', 'Относительная погрешность')
ax.fig.suptitle('Зависимость относительной погрешности')
plt.tight_layout()
plt.savefig('pics/rel_error.png')
```
